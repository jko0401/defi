```python
from random import randint
from time import sleep

for i, row in txns.iterrows():
    txns.loc[txns.index == i, 'Received Cost Basis (USD)'] = dft.geckoPriceAt(row['received_id'], 'usd', row['day'])
    txns.loc[txns.index == i, 'Sent Cost Basis (USD)'] = dft.geckoPriceAt(row['sent_id'], 'usd', row['day'])

    sleep(randint(3,5))
```

```python
tokens = pd.read_csv('tokens_db.csv')
mapping = dict(tokens[['symbol', 'nameid']].values)
mapping['USD'] = 'usd'
```

```python
test = pd.read_csv('test-txns.csv')
```

```python
test['received_id'] = test['Received Currency'].map(mapping)
test['sent_id'] = test['Sent Currency'].map(mapping)
test['fee_id'] = test['Fee Currency'].map(mapping)
```

```python
cbpro = Wallet('Coinbase Pro')
cb = Wallet('Coinbase')
binance = Wallet('Binance')
ledger = Wallet('Ledger')
metamask = Wallet('Metamask')
tempw = Wallet('tempw')

wallets = {'Coinbase Pro':cbpro, 'Coinbase':cb, 'Binance': binance}
```

```python
for i, r in test.iterrows():
    print(i)
    t = pd.to_datetime(r['Date']).strftime('%d-%m-%Y')
    tp = r['Type']
    if tp == 'Buy' or tp == 'Trade' or tp == 'Sell':
        w = wallets[r['Received Wallet']]
        b = (r['received_id'],r['Received Quantity'])
        s = (r['sent_id'],r['Sent Quantity'])
        if r['Fee Amount']:
            f = (r['fee_id'], r['Fee Amount'])
        else:
            f = (0, 0)
        Transaction.transact(t, w, b, s, f, tp)
    elif tp == 'Send':
        w = wallets[r['Sent Wallet']]
        tk = r['sent_id']
        amt = r['Sent Quantity']
        if r['Fee Amount']:
            f = (r['fee_id'], r['Fee Amount'])
        else:
            f = (0, 0)
        Transaction.send(t, w, tk, amt, f)
    elif tp == 'Receive':
        w = wallets[r['Received Wallet']]
        tk = r['received_id']
        amt = r['Received Quantity']
        Transaction.receive(t, w, tk, amt)
    elif tp == 'Transfer':
        o = wallets[r['Sent Wallet']]
        d = wallets[r['Received Wallet']]
        tk = r['sent_id']
        a = r['Sent Quantity']
        if r['Fee Amount']:
            f = (r['fee_id'], r['Fee Amount'])
        else:
            f = (0, 0)
        Transaction.transfer(t, o, d, tk, a, f)
    else:
        print(f'{tp} not supported.')

    
```
