```python
import pandas as pd
from main import Wallet, Coin, Transaction
import defi_tools as dft
```

```python
txns = pd.read_csv('transactions_HIFO_Universal.csv')
txns.drop(index=len(txns)-1, inplace=True)
txns.tail()
```

```python
tokens = dft.geckoFullList()
tokens.reset_index(inplace=True)
tokens = tokens[['symbol', 'id', 'name']]
tokens.symbol = tokens.symbol.str.upper()
tokens.rename(columns={'id':'nameid'}, inplace=True)
```

```python
mapping = dict(tokens[['symbol', 'nameid']].values)
txns['received_id'] = txns['Received Currency'].map(mapping)
txns['sent_id'] = txns['Sent Currency'].map(mapping)
txns['Date'] = pd.to_datetime(txns['Date'])
txns['day'] = txns['Date'].apply(lambda x: x.strftime('%d-%m-%Y'))
```

```python
txns.head()
```

```python
# txns['Received Cost Basis (USD)'][0] = 1
txns.loc[txns.index == 0, 'Received Cost Basis (USD)'] = 0
txns.head()
```

```python
from random import randint
from time import sleep

for i, row in txns.iterrows():
    txns.loc[txns.index == i, 'Received Cost Basis (USD)'] = dft.geckoPriceAt(row['received_id'], 'usd', row['day'])
    txns.loc[txns.index == i, 'Sent Cost Basis (USD)'] = dft.geckoPriceAt(row['sent_id'], 'usd', row['day'])

    sleep(randint(3,5))
```
