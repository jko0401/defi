```python
import defi_tools as defi
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.coingecko.com/'
page = BeautifulSoup(requests.get(url).content,'lxml')    
```

```python
count = page.find_all('div', {'class':'tw-text-xl'})[3].text
```

```python
likes_df = {}
for page in range(int(int(count)/250)):
    print(page*250)
    if page > int(count):
        break
    else:
        try:
            coins = defi.geckoFullList(page=page, per_page=250)
            for coin in coins.id.values:
                c = url + 'en/coins/' + coin
                c_page = BeautifulSoup(requests.get(c).content)
                found = c_page.find('div', {'class':'tw-block tw-inline-flex tw-py-0.5 tw-h-5 tw-items-center tw-px-2 tw-rounded-md tw-text-xs tw-font-medium tw-bg-gray-100 tw-text-gray-800 tw-mb-1 md:tw-mb-0 dark:tw-text-white dark:tw-bg-white dark:tw-bg-opacity-5'})
                likes = found.text.strip('\n').strip(' ').replace(',', '')[:-17]
                likes_df[coin] = int(likes)
        except Exception as e:
            print(coin)
            print(e)
            
```
