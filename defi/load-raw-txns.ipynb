```python
import pandas as pd
import numpy as np
from main import Wallet, Coin, Transaction
import defi_tools as dft
```

```python
tokens = pd.read_csv('tokens_db.csv')
mapping = dict(tokens[['symbol', 'nameid']].values)
mapping['USD'] = 'usd'
mapping['BTRFLY'] = 'butterflydao'
```

```python
data = pd.read_csv('updated-txns.csv')
```

```python
wallets = list(set(list(data['Received Wallet'].values) + list(data['Sent Wallet'].values)))
wallets.remove(wallets[0])
wallets.sort()
wallets
```

```python
binance = Wallet('Binance')
cb = Wallet('Coinbase')
cbpro = Wallet('Coinbase Pro')
metamask = Wallet('Metamask')
ledger = Wallet('Ledger')


wallets = {'Binance': binance, 'Coinbase':cb, 'Coinbase Pro':cbpro, 'ETH Wallet ...50a3d5':metamask, 'ETH Wallet ...C8859C':ledger, 'Ledger':ledger}
```

```python
data['received_id'] = data['Received Currency'].map(mapping)
data['sent_id'] = data['Sent Currency'].map(mapping)
data['fee_id'] = data['Fee Currency'].map(mapping)
data['Date'] = pd.to_datetime(data['Date']).apply(lambda x: x.strftime('%d-%m-%Y'))
```

```python
for i, r in data.iterrows():
    print(i)
    t = r['Date']
    tp = r['Type'].lower()
    if pd.isnull(r['Fee Amount']): fee = 0
    else: fee = r['Fee Amount'] 
    try:
        if tp == 'buy' or tp == 'sell' or tp == 'trade' or tp == 'transfer':
            
            origin =      (wallets[r['Sent Wallet']], r['sent_id'], r['Sent Quantity'], fee)
            destination = (wallets[r['Received Wallet']], r['received_id'], r['Received Quantity'], fee)
            Transaction.transact(t, origin, destination, tp)

        elif tp == 'send' or tp == 'withdraw':
            origin =      (wallets[r['Sent Wallet']], r['sent_id'], r['Sent Quantity'], fee)
            destination = None
            Transaction.transact(t, origin, destination, tp)

        elif tp == 'receive' or tp == 'deposit':
            origin =      None
            destination = (wallets[r['Received Wallet']], r['received_id'], r['Received Quantity'], fee)
            Transaction.transact(t, origin, destination, tp)
        else:
            pass
            print(f'{tp} not supported.')
    except KeyError:
        print(f'!!! {i} skipped. Please Review!!!')
        pass
```
